## 5. Cleaning Operations

```python
index_to_drop = [6529, 303, 159, 8022]
```

```python
minute_rows = netflix['rating'].str.contains('min', na=False)

# Perform the shift
netflix.loc[minute_rows, 'duration'] = netflix.loc[minute_rows, 'rating']
netflix.loc[minute_rows, 'rating'] = np.nan
```

```python
def clean_data(df):
    return (
        df
        .drop(index=index_to_drop)
        .drop_duplicates()
        .assign(**{
            col: df[col].str.strip()
            for col in df.select_dtypes(include='O').columns
        })
        .rename(columns=str.lower)
        .assign(
            title=lambda df_: (
                df_
                .title
                .replace('Death Note', 'Death Note (2017)')
                .replace('FullMetal Alchemist', 'FullMetal Alchemist (2017)')
                .str.strip()
                .str.title()
            ),
            director=lambda df_: (
                df_
                .director
                .str.strip()
                .str.title()
                .fillna('unknown')
            ),
            cast=lambda df_: (
                df_
                .cast
                .apply(lambda x: [] if pd.isna(x) else x.split(', '))
            ),
            country=lambda df_: (
                df_
                .country
                .str.findall('[A-Za-z]+ [A-Za-z]+ [A-Za-z]+|[A-Za-z]+ [A-Za-z]+|[A-Za-z]+')
            ),
            date_added=lambda df_: pd.to_datetime(df_.date_added, format='mixed'),
            movie_duration=lambda df_: (
                df_
                .duration
                .str.extract(r'([0-9]+) min', expand=False)
                .astype(float)
            ),
            tv_show_seasons=lambda df_: (
                df_
                .duration
                .str.extract(r'([0-9]+) Seasons?', expand=False)
                .astype(float)
            ),
            listed_in=lambda df_: (
                df_
                .listed_in
                .str.split(', ')
            )
        )
        .drop(columns=['duration'])
    )
clean_data(netflix)
```

```python
netflix_cleaned = clean_data(netflix)
```

```python
netflix_cleaned.info()
```

## 7. Export the data

```python
def export_data(data, name):
    file_name = f'{name}.csv'
    file_path = os.path.join(PROJECT_DIR, DATA_DIR, file_name)
    data.to_csv(file_path, index=False)
    return pd.read_csv(file_path).head()
```

```python
export_data(netflix_cleaned, 'netflix_cleaned')
```
